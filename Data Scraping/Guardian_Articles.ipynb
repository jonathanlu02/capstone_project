### Load JSON Files to DF

```python

files = glob.glob('articles\*.json')

df=[]
for file in files:
    f = open(file, 'r')
    a1 = pd.read_json(f)
    a1 = a1.fields
    a1 = a1.to_frame()
    df = a1.append(df)
```

```python
df = pd.DataFrame(df['fields'].values.tolist(), index=df.index).reset_index()
```
